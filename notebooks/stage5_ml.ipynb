```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle

with open('../data/loan_model.pkl', 'rb') as f:
    model = pickle.load(f)

df = pd.read_csv('../data/clean_loan_data.csv')
X = df.drop('default', axis=1)

features = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(features['feature'], features['importance'], color='#3498db')
ax.set_title('Feature Importance — What Drives Loan Default?')
ax.set_xlabel('Importance Score')
ax.invert_yaxis()
plt.tight_layout()
plt.savefig('chart_feature_importance.png')
plt.show()
```
