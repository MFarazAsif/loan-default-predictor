```python
query = """
SELECT 
    COUNT(*) as total_loans,
    SUM(`default`) as total_defaults,
    ROUND(AVG(`default`) * 100, 2) as default_rate_percent
FROM loans
"""
print(run_query(query))
```

```python
query = """
SELECT 
    `default`,
    ROUND(AVG(int_rate), 2) as avg_interest_rate,
    ROUND(AVG(loan_amnt), 2) as avg_loan_amount,
    ROUND(AVG(annual_inc), 2) as avg_income,
    COUNT(*) as total
FROM loans
GROUP BY `default`
"""
print(run_query(query))
```

```python
query = """
SELECT 
    CASE 
        WHEN fico_range_low >= 750 THEN 'A - Excellent'
        WHEN fico_range_low >= 700 THEN 'B - Good'
        WHEN fico_range_low >= 650 THEN 'C - Fair'
        WHEN fico_range_low >= 600 THEN 'D - Poor'
        ELSE 'F - Very Poor'
    END as credit_grade,
    COUNT(*) as total_loans,
    ROUND(AVG(`default`) * 100, 2) as default_rate_percent
FROM loans
GROUP BY credit_grade
ORDER BY default_rate_percent DESC
"""
print(run_query(query))
```

```python
query = """
SELECT *
FROM loans
WHERE int_rate > 20
AND fico_range_low < 650
AND dti > 30
ORDER BY int_rate DESC
LIMIT 10
"""
print(run_query(query))
```
